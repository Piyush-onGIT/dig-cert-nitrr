```python
from PIL import Image
import cv2
import numpy as np
```

```python
img = Image.open(r"test.jpg")
sign = Image.open(r"sign.jpeg")

def removebg(sign_image):
  sign_cv2 = np.array(sign_image)

  lower_white = np.array([200, 200, 200], dtype=np.uint8)
  upper_white = np.array([255, 255, 255], dtype=np.uint8)

    # Convert the image to grayscale
  mask = cv2.inRange(sign_cv2, lower_white, upper_white)

    # Invert the mask to get the non-white areas
  mask = cv2.bitwise_not(mask)

  # Apply the mask to the image
  result = cv2.bitwise_and(sign_cv2, sign_cv2, mask=mask)

  # Convert the result back to a PIL Image
  result_pil = Image.fromarray(result)

  return result_pil
```

```python
back_im = img.copy()
sign = sign.resize((300, 200))
x = back_im.width - sign.width
y = back_im.height - sign.height

sign_no_bg = removebg(sign)
sign_no_bg

# back_im.paste(sign, (x, y))

# img.size
# back_im
```
